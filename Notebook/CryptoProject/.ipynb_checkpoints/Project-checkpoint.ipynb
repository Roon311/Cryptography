```python
def initialization(num_octet:int):
    iv = b'\x00' * (num_octet)
    #print(len(iv))
    #print('---------------------------initialization vector-----------------\n')
    #print(f'{iv} \n')
    #print('-----------------------------------------------------------------\n')
    return iv

def is_pkcs7_padded(data):
    last_byte = data[-1]
    if last_byte > 16:
        return False
    for i in range(len(data) - last_byte, len(data)):
        if data[i] != last_byte:
            return False
    return True

def getText(path):
    with open(path,'r', encoding='utf8') as file:
        Text = file.read()
    return Text

def WriteResults(filename):
    filenamee = filename+".txt"
    #file.write()
    #file.close()
    try:
        file = open(filename, "r+")  # open file for reading and writing
    except FileNotFoundError:
        file = open(filename, "w+")  # create file if it doesn't exist
    return file

def drawer(originaltext,ciphertext,encryption_time,plaintext,decryption_time):
    print('------------------------------------------- Original Text --------------------------------------------\n')
    print(f'Original Text:\n\n {originaltext}')
    print('\n----------------------------------------- Encryption -----------------------------------------------\n')
    print(f'\u2022Cipher Text:\n\n{ciphertext}\n')
    print(f'\u2022Encryption Time:  {encryption_time} seconds')
    print('\n----------------------------------------- Decryption -----------------------------------------------\n')
    print(f'\u2022Decrypted Text:\n\n{plaintext}')
    print(f'\u2022Decryption Time:  {decryption_time} seconds')
    print('--------------------------------------------------------------------------------------------------------')
```

# PGP-CFB Mode

```python
def encrypt_PGP_CFB_Mode(plaintext,key, length=128):
    assert length % 8 == 0
    num_octet = length // 8
    BS = num_octet
    ciphertext = b''
    plaintext = plaintext.encode(encoding='utf8')  # Convert data to bytes
    iv = initialization(num_octet)
    fr = iv
    cipher = AES.new(key, AES.MODE_CFB, iv=iv)
    prefix = os.urandom(num_octet + 2)
    fre = cipher.encrypt(fr)
    C=bytes([f ^ b for f, b in zip(fre,prefix[0:BS])])
    ciphertext +=C
    fr=C#Load FR with C[1] through C[BS]
    fre = cipher.encrypt(fr)


    C_bs_plus_1_2 = bytes([f ^ prefix[BS+i] for i, f in enumerate(fre[:2])])
    ciphertext+=C_bs_plus_1_2
   # print('-------------------------')
   # print(prefix)
   # print('-------------------------')

    # resynchronization
    fr = fr[2:] + C_bs_plus_1_2#FR is loaded with C[3] through C[BS+2]
    while plaintext:
        if(len(plaintext)<BS):
            plaintext=pad(plaintext,BS)
        data=plaintext[:BS]
        fre = cipher.encrypt(fr)
        ciphertext_block = bytes([f ^ b for f, b in zip(fre, data)])
        ciphertext+=ciphertext_block
        fr=ciphertext_block
        plaintext = plaintext[BS:]
    return ciphertext

def decrypt_PGP_CFB_Mode(ciphertext,key,length=128):
    assert length % 8 == 0
    num_octet = length // 8
    BS = num_octet
    iv = initialization(num_octet)
    fr = iv
    cipher = AES.new(key, AES.MODE_CFB, iv=iv)
    fre = cipher.encrypt(fr)
    prefix = bytes([f ^ b for f, b in zip(fre, ciphertext[0:16])])     
    fr=ciphertext[0:16]
    fre = cipher.encrypt(fr)
    prefix2 = bytes([f1 ^ c1 for f1, c1 in zip(fre[:2],ciphertext[16:18])])
    prefix+=prefix2         # I am having an issue with the prefix
    
   # print('-------------------------')
   # print(prefix)
   # print('-------------------------')

    fr=ciphertext[2:18]
    fre = cipher.encrypt(fr)
    c=ciphertext[18:34]
    plaintext= bytes([f ^ b for f, b in zip(fre, c)])
    ciphertext=ciphertext[34:]
    fr=c
    while ciphertext:
        data=ciphertext[:BS]
        fre = cipher.encrypt(fr)
        plaintext_block = bytes([f ^ b for f, b in zip(fre, data)])
        plaintext+=plaintext_block
        fr=data
        ciphertext=ciphertext[BS:]
    if is_pkcs7_padded(plaintext)==True:
        plaintext=unpad(plaintext,BS)
    return plaintext.decode(encoding='utf8')

def tester(path,key= get_random_bytes(16)):
    Text=getText(path)
    start = time.time()
    ciphertext = encryptor(key,Text)
    end =time.time()
    encryption_time=end-start
    start = time.time()
    plaintext=decryptor(key, ciphertext)
    end =time.time()
    decryption_time=end-start
    return Text,ciphertext,encryption_time,plaintext,decryption_time
```

# Unicode To Ascii

```python
# Taken from https://gist.github.com/tushortz/9fbde5d023c0a0204333267840b592f9
def unicodetoascii(text):

    TEXT = (text.
            replace('\\xe2\\x80\\x99', "'").
            replace('\\xc3\\xa9', 'e').
            replace('\\xe2\\x80\\x90', '-').
            replace('\\xe2\\x80\\x91', '-').
            replace('\\xe2\\x80\\x92', '-').
            replace('\\xe2\\x80\\x93', '-').
            replace('\\xe2\\x80\\x94', '-').
            replace('\\xe2\\x80\\x94', '-').
            replace('\\xe2\\x80\\x98', "'").
            replace('\\xe2\\x80\\x9b', "'").
            replace('\\xe2\\x80\\x9c', '"').
            replace('\\xe2\\x80\\x9c', '"').
            replace('\\xe2\\x80\\x9d', '"').
            replace('\\xe2\\x80\\x9e', '"').
            replace('\\xe2\\x80\\x9f', '"').
            replace('\\xe2\\x80\\xa6', '...').#
            replace('\\xe2\\x80\\xb2', "'").
            replace('\\xe2\\x80\\xb3', "'").
            replace('\\xe2\\x80\\xb4', "'").
            replace('\\xe2\\x80\\xb5', "'").
            replace('\\xe2\\x80\\xb6', "'").
            replace('\\xe2\\x80\\xb7', "'").
            replace('\\xe2\\x81\\xba', "+").
            replace('\\xe2\\x81\\xbb', "-").
            replace('\\xe2\\x81\\xbc', "=").
            replace('\\xe2\\x81\\xbd', "(").
            replace('\\xe2\\x81\\xbe', ")")

                 )
    return TEXT
```

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes

def encrypt_Standard_CFB_mode(key, iv, plaintext):
    
    cipher = AES.new(key, AES.MODE_CFB, iv=iv)
    
    ciphertext = cipher.encrypt((pad(plaintext, AES.block_size))) #cipher.encrypt(plaintext)
    return ciphertext


def decrypt_Standard_CFB_mode(key, iv, ciphertext):
    decipher = AES.new(key, AES.MODE_CFB, iv=iv)
    decryptedtext = unpad(decipher.decrypt(ciphertext), AES.block_size) # decipher.decrypt(ciphertext)  #
    return decryptedtext

key = get_random_bytes(16)
iv = get_random_bytes(16)

data = bytes("Message", encoding = "UTF-8")

start = time.time()
encrypted_data = encrypt_Standard_CFB_mode(key, iv, data)
decrypted_data = decrypt_Standard_CFB_mode(key, iv, encrypted_data)
end = time.time()

print('Original message:', str(data)[2:-1] )
print('Encrypted message:', str(encrypted_data)[2:-1] )
print('Decrypted message:', str(decrypted_data)[2:-1] )
print()
print("It takes {}".format(end - start))
```

# Comparison between PGP_CFB Mode and Standard CFB Mode

```python
def ComparePGP_CFB(SecurityLevel,List,AverageOver):
    List_PGP_CFB_Both = {}
    List_PGP_CFB_Enc = {}
    List_PGP_CFB_Dec = {}
    KeySize = int(SecurityLevel/8)
    List = [1,5,10,100]
    for i in List:
        filename = str(i)+"KB_file"
        with open(f"../Project/InputFiles/{filename}.txt",'r', encoding='utf8') as file:
            Text = file.read()
        plaintext = Text #bytes(Text, encoding='utf8')

        key = get_random_bytes(KeySize)  # os.urandom(16)
        Average_Times_Over_File_Both = []
        Average_Times_Over_File_Enc = []
        Average_Times_Over_File_Dec = []
        for j in range(AverageOver): 
            start = time.time()
            ciphertext = encrypt_PGP_CFB_Mode(plaintext, key)
            mid = time.time()
            decrypted_plaintext = decrypt_PGP_CFB_Mode(ciphertext, key)
            end = time.time()
            Average_Times_Over_File_Both.append(end - start)
            Average_Times_Over_File_Enc.append(mid - start)
            Average_Times_Over_File_Dec.append(end - mid)
        #plaintext = unicodetoascii( str(plaintext) ).replace("\\n","\n")[2:-1]
        #decrypted_plaintext = unicodetoascii( str(decrypted_plaintext) ).replace("\\n","\n")[2:-1]
        List_PGP_CFB_Both[filename] = (1000.0*sum(Average_Times_Over_File_Both)) /len(Average_Times_Over_File_Both)
        List_PGP_CFB_Enc[filename] = (1000.0*sum(Average_Times_Over_File_Enc)) /len(Average_Times_Over_File_Enc)
        List_PGP_CFB_Dec[filename] = (1000.0*sum(Average_Times_Over_File_Dec)) /len(Average_Times_Over_File_Dec)
        
    return List_PGP_CFB_Both, List_PGP_CFB_Enc, List_PGP_CFB_Dec
```

```python
def CompareStandard_CFB(SecurityLevel,List,AverageOver):
    List_Standard_CFB_Both = {}
    List_Standard_CFB_Enc = {}
    List_Standard_CFB_Dec = {}
    KeySize = int(SecurityLevel/8)
    List = [1,5,10,100]
    for i in List:
        filename = str(i)+"KB_file"
        with open(f"../Project/InputFiles/{filename}.txt",'r', encoding='utf8') as file:
            Text = file.read()
        plaintext = bytes(Text, encoding='utf8')

        key = get_random_bytes(KeySize)
        iv = get_random_bytes(16)                 #----> make sure about this
        Average_Times_Over_File_Both = []
        Average_Times_Over_File_Enc = []
        Average_Times_Over_File_Dec = []
        for j in range(30):
            start = time.time()
            ciphertext = encrypt_Standard_CFB_mode(key, iv, plaintext)
            mid = time.time()
            decrypted_plaintext = decrypt_Standard_CFB_mode(key, iv, ciphertext)
            end = time.time()
            Average_Times_Over_File_Both.append(end - start)
            Average_Times_Over_File_Enc.append(mid - start)
            Average_Times_Over_File_Dec.append(end - mid)
        #plaintext = unicodetoascii( str(plaintext) ).replace("\\n","\n")[2:-1]
        #decrypted_plaintext = unicodetoascii( str(decrypted_plaintext) ).replace("\\n","\n")[2:-1]
        List_Standard_CFB_Both[filename] = (1000.0*sum(Average_Times_Over_File_Both)) /len(Average_Times_Over_File_Both)
        List_Standard_CFB_Enc[filename] = (1000.0*sum(Average_Times_Over_File_Enc)) /len(Average_Times_Over_File_Enc)
        List_Standard_CFB_Dec[filename] = (1000.0*sum(Average_Times_Over_File_Dec)) /len(Average_Times_Over_File_Dec)
        
    return List_Standard_CFB_Both, List_Standard_CFB_Enc, List_Standard_CFB_Dec
```

```python
from Crypto.Util.number import inverse, getPrime, bytes_to_long, long_to_bytes
from Crypto.Random.random import randrange

# Key generation
def generate_key(bit_length):
    p = getPrime(bit_length // 2)
    q = getPrime(bit_length // 2)
    n = p * q
    phi = (p - 1) * (q - 1)
    e = 65537  # commonly used public exponent
    d = pow(e, -1, phi)
    return ((e, n), (d, n))

# Encryption
def encrypt_RSA(plaintext, public_key):
    e, n = public_key
    plaintext_int = bytes_to_long(plaintext)  # converts from bytes to long integers As RSA is working with numbers
    ciphertext_int = pow(plaintext_int, e, n)  # RSA Encryption formula
    return long_to_bytes(ciphertext_int)    # converts from integers back to bytes to be represented

# Decryption
def decrypt_RSA(ciphertext, private_key):
    d, n = private_key
    ciphertext_int = bytes_to_long(ciphertext)  # converts from bytes to long integers As RSA is working with numbers
    plaintext_int = pow(ciphertext_int, d, n)  # RSA Decryption formula
    return long_to_bytes(plaintext_int)   # converts from integers back to bytes to be represented


# Example usage
SecurityLevel = 256
plaintext = b'Hello, world!'
bit_length = 2048
public_key, private_key = generate_key(bit_length)
ciphertext = encrypt_RSA(plaintext, public_key)
decrypted_plaintext = decrypt_RSA(ciphertext, private_key)

print("Original Text: {}".format(plaintext))
print("Encrypted Text: {}".format(ciphertext))
print("Decrypted Text: {}".format(decrypted_plaintext))
```
